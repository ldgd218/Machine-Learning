```python

import requests
from bs4 import BeautifulSoup as BS
import json
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_national_capitals_in_alphabetical_order'
```

```python
data = requests.get(url)
print data
```

```python
print data.text
```

```python
soup = BS(data.text) #BS is constructor which creates data treat tree 
print type(soup)
```

```python
print soup
```

```python
table = soup.find_all('table') #recursively searches for tables in the above code(even if nested tables-->treat as different tables)
print len(table), type(table)
```

```python
for ix in range(len(table)):
    print len(str(table[ix]))
```

```python
t = table[2]
print str(t)[:50]
```

```python
tr = t.find_all('tr')
print len(tr)
```

```python
country = []
for ix in range(1, len(tr)):
    
    c_data = {}
    
    #city
    td = tr[ix].find_all('td')
    a = td[0].find_all('a')[0]
    #print "City: ", a.text, "| Country:",
    #print a['href'] #href is a special prop which cannot be accessed using '.' So we use it like dictionary
    c_data['city'] = a.text
    
    #country
    n = td[1].find_all('a')[0]
    #print n.text
    c_data['country'] = n.text
    country.append(c_data)
    #print '-----------------------'
```

```python
len(country)
```

```python
#json = javascript object notation
#we store in form of json in form of key value pairs
```

```python
data = {'data' : country}
```

```python
y = json.dumps(data)
print len(y)

f = open('countries_and_caps.json', 'w')
f.write(y)
f.close()
```
